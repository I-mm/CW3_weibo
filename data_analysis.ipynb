#### （1）Find the content posted by the people who have more than 1000 followers

```python
sql_t1 = "SELECT DISTINCT weibo_user.user_id, weibo_user.user_name, content " \
         "FROM weibo_user, " \
         "user_post " \
         "WHERE weibo_user.follower_num > 1000 " \
         "AND weibo_user.user_id = user_post.poster_id; "
cursor.execute(sql_t1)
rows = cursor.fetchall()
df = pd.DataFrame([[ij for ij in i] for i in rows])
df.rename(columns={0:'user_id',1:'user_name', 2:'content'},inplace=True)
print("TASK1: Content posted by the people who have more than 1000 followers")
df
```

#### （2）Research the law and features of the content which post_number is more than 5.

```python
sql_t2 = "SELECT DISTINCT weibo_user.user_id, weibo_user.user_name, content " \
         "FROM weibo_user, " \
         "user_post " \
         "WHERE weibo_user.user_id = user_post.poster_id " \
         "AND weibo_user.post_num > 5; "
cursor.execute(sql_t2)
rows = cursor.fetchall()
df = pd.DataFrame([[ij for ij in i] for i in rows])
df.rename(columns={0:'user_id',1:'user_name', 2:'content'},inplace=True)
print("TASK2: Law and features of the content which post_number is more than 5")
df
```

#### （3）Please find the spammer and advertiser.

```python
sql_t3_1 = "SELECT DISTINCT user_id, user_name " \
           "FROM weibo_user " \
           "WHERE weibo_user.is_spammer = 1; "
cursor.execute(sql_t3_1)
rows = cursor.fetchall()
df = pd.DataFrame([[ij for ij in i] for i in rows])
df.rename(columns={0:'user_id',1:'user_name'},inplace=True)
print("TASK3_1: Spammer")
df
```

```python
sql_t3_2 = "select DISTINCT user_id, user_name " \
           "from user_post, weibo_user " \
           "where weibo_user.user_id = user_post.poster_id and " \
           "(user_post.content like '%微信%' " \
           "or user_post.content like '%V信%' " \
           "or user_post.content like '%weixin%' " \
           "or user_post.content like '%QQ%' " \
           "or user_post.content like '%价格%' " \
           "or user_post.content like '%联系%' " \
           "or user_post.content like '%包邮%'); "
cursor.execute(sql_t3_2)
rows = cursor.fetchall()
df = pd.DataFrame([[ij for ij in i] for i in rows])
df.rename(columns={0:'user_id',1:'user_name'},inplace=True)
print("TASK3_2: Advertiser")
df
```
